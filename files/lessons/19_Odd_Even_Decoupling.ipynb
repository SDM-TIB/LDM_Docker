#Odd-Even Decoupling

What is odd-even decoupling?  Let's pull up a video from the Burgers' Equation exercise and take a look.  


```python
from IPython.display import HTML
from base64 import b64encode
video = open("../videos/laxfricfl1.mp4", "rb").read()
video_encoded = b64encode(video)
video_tag = '<video controls alt="test" src="data:video/x-m4v;base64,{0}">'.format(video_encoded)
HTML(data=video_tag)
```

Notice how the points along the leading edge of the wave behave?  They cascade down in pairs.  This is not an evanescent behavior, it will continue for as long as the simulation is run, so why does it happen?  

The particular video shown above is using the Lax-Friedrichs scheme in which

$$u^{n+1}_i = f \left(u^n_{i-1}, u^n_{i+1} \right)$$

The only points on the grid that $u^{n+1}_i$ depends on are those parent points, $u^n_{i-1}$ and $u^n_{i+1}$.  

If we drew out a little scheme diagram for this, it might look like the following.  

```python
from IPython.display import Image
Image('un+1.png')
```

What about if we did the same for $u^n_i$? By extension, its parents should be $u^{n-1}_{i-1}$ and $u^{n-1}_{i+1}$.

```python
Image('un.png')
```

Above we can see that $u^{n+1}_i$ doesn't depend on the value of $u^n_i$.  Which means it also doesn't depend on either $u^{n-1}_{i-1}$ or $u^{n-1}_{i+1}$.  

If we connect all of the points on the grid that *do* depend on one another, you get something that looks like this:

```python
Image('oddevengrid.png')
```

###What does this all mean?

When certain schemes are used which depend only on $u^n_{i-1}$ and $u^n_{i+1}$, the problem is solved twice, in tandem, with each solution offset from the other by one step in space.  This doesn't mean that it's right or wrong, but it is something to be aware of when choosing a scheme for a particular problem.
